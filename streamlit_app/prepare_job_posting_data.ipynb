```python
import os
import pandas as pd
import numpy as np
import ast

# Paths
build_project_path = os.environ.get(
    "BUILD_PROJECT_PATH",
    r"C:\Users\L110006\OneDrive - Eli Lilly and Company\personal files\build project\fine-tuning-build-project"
)
streamlit_data_path = os.path.join(build_project_path, "streamlit_app", "data")
parquet_file = os.path.join(streamlit_data_path, 'job_postings.parquet')

# Read the Parquet file
job_postings_df = pd.read_parquet(parquet_file, engine="pyarrow")

# --- Extract first URL from stringified list ---
def extract_first_url(url_str):
    try:
        url_list = ast.literal_eval(url_str)
        return url_list[0] if url_list else None
    except:
        return None

job_postings_df['url'] = job_postings_df['url'].apply(extract_first_url)

# --- Convert posted date to datetime ---
job_postings_df['date_posted'] = pd.to_datetime(job_postings_df['date_posted'], errors='coerce')

# --- Rename columns ---
job_postings_df.rename({
    'COMPANY_NAME': 'company',
    'TITLE_RAW': 'job_posting_title',
    'ONET_2019': 'onet_code',
    'ONET_2019_NAME': 'onet_name'
}, axis=1, inplace=True)

# --- Optional: sample fraction ---
job_postings_df = job_postings_df.sample(frac=(1/6), random_state=42).reset_index(drop=True)

# Check the DataFrame
job_postings_df.head()
```
